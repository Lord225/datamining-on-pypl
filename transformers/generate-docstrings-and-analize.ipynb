```python
import pandas as pd
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
MODEL = 'google/pegasus-cnn_dailymail'

model = transformers.AutoModelForSeq2SeqLM.from_pretrained('F:/pyrepos/datamining-on-pypl/transformers/models/checkpoint-150')

model = model.to(device)

tokenizer = transformers.AutoTokenizer.from_pretrained(MODEL)
```

```python
# generate docstrings for functions without docstrings

from tqdm import tqdm

no_docstrings = functions[functions.docstring.isna()]

no_docstrings
```

```python
def prep_data_body(row):
    func_name = row['name']
    func_args = row['args']
    func_body = row['body']

    def format_args(args):
        return ', '.join(args.replace('{', '').replace('}', '').split())
    
    def format_body(body):
        return body.replace('\n', ' ').replace('\r', ' ')

    # rebuild function
    func = f"def {func_name}({format_args(func_args)}):\n {format_body(func_body)}"

    return func

no_docstrings['function'] = no_docstrings.apply(prep_data_body, axis=1)
```

```python
no_docstrings['function']
```

```python
tokenized = tokenizer(no_docstrings['function'][:2000].tolist(), return_tensors='pt', padding=True, truncation=True)
```

```python
batch_size = 2
lenght = len(tokenized['input_ids'])
outputs = []
tq = tqdm(range(0, lenght, batch_size))
for i in tq:
    batch_input_ids = tokenized['input_ids'][i:i+batch_size].to(device)
    batch_attention_mask = tokenized['attention_mask'][i:i+batch_size].to(device)
    generated_ids = model.generate(
        input_ids=batch_input_ids,
        attention_mask=batch_attention_mask,
        max_length=150,
        length_penalty=1,
        early_stopping=True 
    )

    decoded = [tokenizer.decode(g, skip_special_tokens=True) for g in generated_ids]

    outputs.extend(decoded)

    # set description to generated docstring
    tq.set_description(decoded[0])
    
```

```python
# save to db
indexes = no_docstrings['body'][:2000].index
indexes
```

```python
outputs_df = pd.DataFrame(outputs, columns=['docstring']).set_index(indexes)

# save to db
outputs_df.to_sql('functions_docstrigns_generated', db, if_exists='replace')
# commit
db.commit()
```
